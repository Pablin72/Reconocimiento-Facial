# **Introducción**

**Sobre el conjunto de datos**
---
---

Es una **colección de imágenes faciales** de varios **celebridades y figuras públicas**, incluyendo **actores, músicos, políticos, deportistas e influenciadores de las redes sociales**. El conjunto de datos incluye un total de **107,818 imágenes de 1,063 individuos diferentes**.

Las imágenes se proporcionan en formato **JPEG** y tienen **tamaños y calidades variables**. Algunas imágenes son **retratos de estudio de alta calidad**, mientras que otras son **fotografías espontáneas tomadas en diferentes entornos**. Las imágenes también varían en términos de iluminación, poses, expresiones faciales y otros factores.

Además de las imágenes, el conjunto de datos también **incluye un archivo CSV** que contiene **metadatos** para **cada individuo, incluyendo su nombre, género** y el **número de imágenes disponibles para ese individuo**.

El conjunto de datos está destinado a ser utilizado en investigaciones relacionadas con el **reconocimiento facial y la visión por computadora**, y se podría utilizar para **entrenar algoritmos de aprendizaje automático** para tareas como el **reconocimiento facial**, el **análisis de expresiones faciales** y la **estimación de la edad**.

**Problema a abordar**
---

---
El **problema de reconocimiento facial** en el **contexto de este conjunto de datos** consiste en desarrollar un **modelo de aprendizaje automático** que pueda **identificar con precisión a las personas** a partir de sus **imágenes faciales**. Dada una nueva imagen de una cara de una persona, el **objetivo** del **modelo** sería **identificar correctamente** a la **persona** en la **imagen**, comparándola con las **imágenes de individuos conocidos en el conjunto de datos**.

El **problema de reconocimiento facial** es **desafiante** debido a la **variabilidad de las características faciales** entre los **individuos** y la **presencia de diversos factores** que pueden **afectar la apariencia facial**, como la **iluminación, la pose, la expresión y las obstrucciones**. Sin embargo, el **tamaño grande** y la **diversidad** del **conjunto de datos de reconocimiento facial de PINS** brindan una **buena oportunidad** para **desarrollar y probar modelos de reconocimiento facial** que sean **robustos frente a estos factores**.

El **problema del reconocimiento facial** tiene muchas **aplicaciones potenciales**, incluyendo **sistemas de seguridad, control de acceso, identificación personal** y más. Sin embargo, es **importante** asegurarse de que estos **sistemas estén diseñados y utilizados de manera ética**, con las salvaguardias adecuadas para proteger la **privacidad y prevenir el mal uso**.

**Marco de solución**
---
---


El modelo **FaceNet** se puede utilizar para **resolver el problema de reconocimiento facial** mediante el aprendizaje de una **incrustación de alta dimensionalidad** de **imágenes faciales** que **captura las características únicas** de

 la **cara de cada individuo**. Aquí hay un resumen **general** de cómo se puede utilizar **FaceNet** para resolver este problema:

* **Preprocesamiento de los datos**: El primer paso es preprocesar los datos normalizando los valores de píxeles y alineando las caras en las imágenes. Esto se puede hacer utilizando un algoritmo de detección de caras y un algoritmo de alineación de caras.

* **Entrenamiento del modelo FaceNet**: Una vez que los datos han sido preprocesados, el siguiente paso es entrenar el modelo FaceNet. Durante el entrenamiento, el modelo aprende a asignar imágenes faciales a un espacio de incrustación de alta dimensionalidad donde las caras similares están cerca y las caras diferentes están lejos.

* **Ajuste fino del modelo FaceNet**: Después de entrenar el modelo, se puede ajustar fino en el conjunto de datos de reconocimiento facial de PINS para mejorar su rendimiento en este conjunto de datos específico. Esto implica utilizar transferencia de aprendizaje para adaptar el modelo preentrenado al conjunto de datos de PINS o entrenar el modelo con datos adicionales del conjunto de datos.

* **Reconocimiento facial**: Una vez entrenado el modelo, se puede utilizar para reconocer caras en nuevas imágenes. Dada una nueva imagen de una cara, el modelo produce su vector de incrustación, que se compara con los vectores de incrustación de individuos conocidos en el conjunto de datos. La persona cuyo vector de incrustación es más cercano al vector de entrada se reconoce como la persona en la imagen.

* **Evaluación**: El rendimiento del sistema de reconocimiento facial se puede evaluar utilizando métricas como precisión, exactitud, recuperación y puntuación F1. También se puede utilizar validación cruzada y pruebas en un conjunto de validación separado para asegurarse de que el modelo no se esté ajustando demasiado a los datos de entrenamiento.

En general, **FaceNet proporciona** una **solución efectiva** para el **reconocimiento facial** mediante el **aprendizaje de una representación de alta dimensionalidad** de **imágenes faciales** que **puede capturar las características únicas** de la **cara de cada individuo**. El modelo se puede entrenar en **conjuntos de datos a gran escala**, como el **conjunto de datos de reconocimiento facial de PINS**, para lograr una **alta precisión y robustez** en el **reconocimiento de caras** en una **variedad de entornos**.

En este caso, estaremos utilizando un **modelo FaceNet preentrenado**. Utilizar un **modelo FaceNet preentrenado** puede ser un enfoque **muy efectivo** para las tareas de **reconocimiento facial**. Los **modelos preentrenados** son modelos que ya han sido entrenados con una gran cantidad de datos y han aprendido a reconocer una amplia variedad de características importantes para el reconocimiento facial. Al utilizar un **modelo preentrenado**, se puede aprovechar el conocimiento que el modelo ya ha adquirido durante el proceso de entrenamiento y ahorrar tiempo y recursos que se necesitarían para entrenar el modelo desde cero.

**FaceNet** es un modelo **muy poderoso** que ha sido entrenado en **conjuntos de datos grandes** como el **conjunto de datos CASIA-Web

Face** y ha logrado un **rendimiento de vanguardia** en **varios conjuntos de datos de referencia para el reconocimiento facial**. Utilizar un **modelo FaceNet preentrenado** puede proporcionar un buen punto de partida para construir un **sistema de reconocimiento facial** y puede ahorrar mucho tiempo y esfuerzo en comparación con el entrenamiento del modelo desde cero.

# **Carga de datos**
---

Ahora vamos a **cargar los datos en la memoria**, un **paso crucial** en el **reconocimiento facial**. En este problema, típicamente trabajamos con **dos tipos de conjuntos de datos**.

* El **primero** es la **colección de imágenes** que utilizaremos para **probar las capacidades de inferencia de nuestro modelo**.

* El **segundo** es la **base de datos**, que es una **colección de imágenes** utilizadas para **calcular las incrustaciones de caras**. Luego, comparamos estas incrustaciones con las incrustaciones de las nuevas imágenes para realizar el **reconocimiento facial**, es decir, el **proceso de determinar si dos imágenes representan a la misma persona**. Este proceso se conoce como **verificación facial**, y una **versión más compleja** de esto es el **reconocimiento facial**, que implica **reconocer a un individuo** dentro de un **conjunto de múltiples individuos**.

La **base de datos** tiene un **impacto significativo** en el **rendimiento de nuestro modelo** de **dos maneras**.

* En primer lugar, la **calidad de las imágenes** en la **base de datos** puede afectar significativamente la **capacidad de nuestro modelo** para **reconocer caras**. Si las imágenes en la **base de datos contienen las características faciales máximas** del **individuo**, nuestro modelo **reconocerá mejor a esa persona**. Por el contrario, si **recopilamos imágenes al azar**, el **rendimiento de nuestro modelo puede verse afectado**. Sin embargo, **seleccionar manualmente** la **base de datos** es **a menudo impráctico debido a su tamaño**. En esos casos, podemos usar **imágenes aleatorias en la base de datos**.

* El **segundo factor** que **afecta el rendimiento del modelo** es el **número total de imágenes** que tenemos **por individuo**. En este **conjunto de datos**, tenemos un **gran número de imágenes**, y las **imágenes varían significativamente** en términos de **iluminación, pose y expresión**. Para tener en cuenta esta **diversidad**, necesitamos una **base de datos** lo suficientemente **grande como para incluir incrustaciones de varias imágenes diversas**. Por lo tanto, utilizaremos **10 imágenes por individuo** para asegurarnos de que nuestra base de datos sea lo suficientemente **completa y diversa como para mejorar el rendimiento del modelo**.

```python
# Obtenemos el directorio
root_path = '/content/drive/MyDrive/Colab Notebooks/Reconocimiento Facial/Fotos'

# Recolectamos los nombres de las personas
dir_names = os.listdir(root_path)
person_names = [name.split("_")[-1].title() for name in dir_names]
n_individuals = len(person_names)

print(f"Total de individuos: {n_individuals}\n")
print(f"Nombre de los individuos : \n\t{person_names}")
```

```python
# Cantidad de imagenes por personas
n_images_per_person = [len(os.listdir(root_path + '/' + name)) for name in dir_names]
n_images = sum(n_images_per_person)

# Show
print(f"Total Number of Images : {n_images}.")
```

```python
# Plot the Distribution of number of images per person.
fig = px.bar(x=person_names, y=n_images_per_person, color=person_names)
fig.update_layout({'title':{'text':"Fotos por persona"}})
fig.show()
```

```python
# Select all the file paths : 50 images per person.
filepaths = [path  for name in dir_names for path in glob(root_path + "/" + name + "/*")[:500]]
print(filepaths)
np.random.shuffle(filepaths)
print(f"Total number of images to be loaded : {len(filepaths)}")

# Create space for the images
all_images = np.empty(shape=(len(filepaths), IMG_W, IMG_H, IMG_C), dtype = np.float32)
all_labels = np.empty(shape=(len(filepaths), 1), dtype = np.int32)

```

```python
import random

# Crear un arreglo ordenado del 0 al 3499
aleatorios = list(range(0, 3000))

# Desordenar los elementos del arreglo
random.shuffle(aleatorios)

print(aleatorios)
```

```python
#MODIFICAR ESTE CODIGO PARA QUE LAS IMAGENES SE GUARDEN DE MANERA ALEATORIA, IR EN ORDEN CON UN CONTADOR RECORRIENDO UN ARREGLO QUE TIENE LOS NUMEROS
#SIN ORDEN, POR EJEMPLO 23, 6, 98, 123, 3214 ...
# For each path, load the image and apply some preprocessing.

for index, path in tqdm(enumerate(filepaths), desc="Loading Data"):

    # Extract label
    label = [name for name in dir_names if name in path][0]
    label = person_names.index(label.title())

    # Load the Image
    image = plt.imread(path)

    # Convert image stype
    image = image.astype(np.float32)/255.0

    all_images[aleatorios[index]] = image
    all_labels[aleatorios[index]] = label
```

# **Data Visualization**
---

Now that we have **successfully loaded our data set**, we can **proceed to visualize it**. **Visualization** is a **crucial step** in any **data analysis task**, as it helps us **gain insight** into the **data and identify any patterns** or **anomalies that might be present**. In the case of **face recognition**, **visualization allows** us to get a **sense of the quality** of the **images** and how well they **represent each individual**.

```python
#MODIFICAR ESTE CODIGO PARA QUE SOLO RECORRA EL ARREGLO ALEATORIO QUE YA SE CREO
def show_data(
    images: np.ndarray,
    labels: np.ndarray,
    GRID: tuple=(7,4),
    FIGSIZE: tuple=(25,50),
    recog_fn = None,
    database = None
) -> None:

    """
    Function to plot a grid of images with their corresponding labels.

    Args:
        images (numpy.ndarray): Array of images to plot.
        labels (numpy.ndarray): Array of corresponding labels for each image.
        GRID (tuple, optional): Tuple with the number of rows and columns of the plot grid. Defaults to (15,6).
        FIGSIZE (tuple, optional): Tuple with the size of the plot figure. Defaults to (30,50).
        recog_fn (function, optional): Function to perform face recognition. Defaults to None.
        database (dictionary, optional): Dictionary with the encoding of the images for face recognition. Defaults to None.

Returns:
        None
    """

    # Plotting Configuration
    plt.figure(figsize=FIGSIZE)
    n_rows, n_cols = GRID
    n_images = n_rows * n_cols
    contador = 0
    # loop over the images and labels
    for index in range(n_images):

        # Select image in the corresponding label randomly
        image, label = images[index], person_names[int(labels[index])]

        # Create a Subplot
        plt.subplot(n_rows, n_cols, index+1)

        # Plot Image
        plt.imshow(image)
        plt.axis('off')

        if recog_fn is None:
            # Plot title
            plt.title(label)
        else:
            recognized = recog_fn(image, database)
            plt.title(f"True:{label}\nPred:{recognized}")

    # Show final Plot
    plt.tight_layout()
    plt.show()
```

```python
def create_model():
    model = Sequential([
        Conv2D(filters=128, kernel_size=(4, 4), padding='valid', input_shape=(150, 150, 3)),
        Activation('relu'),
        MaxPooling2D(pool_size=(2, 2)),
        BatchNormalization(),

        Conv2D(filters=64, kernel_size=(3, 3), padding='valid', kernel_regularizer=l2(0.1)),
        Activation('relu'),
        MaxPooling2D(pool_size=(2, 2)),
        BatchNormalization(),

        Conv2D(filters=32, kernel_size=(3, 3), padding='valid', kernel_regularizer=l2(0.1)),
        Activation('relu'),
        MaxPooling2D(pool_size=(2, 2)),
        BatchNormalization(),

        Flatten(),


        Dense(units=64, activation='relu'),
        Dense(units=32, activation='relu'),
        Dense(units=16, activation='relu'),
        Dense(units=6, activation='softmax')
    ])

    return model
```

```python
X = all_images
```

###Dummies

```python
# Crear un DataFrame a partir de las etiquetas
y = pd.DataFrame(all_labels, columns=['label'])

# Convertir las etiquetas a variables ficticias (dummies)
y_dummies = pd.get_dummies(y['label'])
```

```python
train_images, test_images, train_labels, test_labels = train_test_split(X, y_dummies, test_size=0.3, random_state=50)
```

```python
cnn_model = create_model()
```

```python
print(cnn_model.summary())
```

```python
cnn_model.compile(optimizer= 'adam', loss='categorical_crossentropy',  metrics=['accuracy'])
```

```python
cnn_model.fit(x= train_images, y = train_labels, validation_data = (test_images, test_labels),
             batch_size=30, epochs=150, verbose = 1)
```

```python
loss = pd.DataFrame(cnn_model.history.history)
plt.figure(figsize = (12, 12))
sns.lineplot(data = loss, lw = 3)
sns.despine()
```

```python
from sklearn.metrics import confusion_matrix, classification_report

# Obtener las predicciones para los conjuntos de entrenamiento y prueba
train_predictions = cnn_model.predict(train_images)
test_predictions = cnn_model.predict(test_images)

# Obtener las clases predichas
train_pred_classes = np.argmax(train_predictions, axis=1)
test_pred_classes = np.argmax(test_predictions, axis=1)

# Obtener las clases verdaderas
train_true_classes = np.argmax(train_labels.values, axis=1)
test_true_classes = np.argmax(test_labels.values, axis=1)

# Generar la matriz de confusión y el informe de clasificación para el conjunto de entrenamiento
train_confusion_matrix = confusion_matrix(train_true_classes, train_pred_classes)
train_classification_report = classification_report(train_true_classes, train_pred_classes)

# Generar la matriz de confusión y el informe de clasificación para el conjunto de prueba
test_confusion_matrix = confusion_matrix(test_true_classes, test_pred_classes)
test_classification_report = classification_report(test_true_classes, test_pred_classes)

# Imprimir la matriz de confusión y el informe de clasificación para el conjunto de entrenamiento
print("Train Confusion Matrix:")
print(train_confusion_matrix)
print("\nTrain Classification Report:")
print(train_classification_report)

# Imprimir la matriz de confusión y el informe de clasificación para el conjunto de prueba
print("Test Confusion Matrix:")
print(test_confusion_matrix)
print("\nTest Classification Report:")
print(test_classification_report)
```

```python
from PIL import Image

def identify_person(image_path, model):
    # Load and preprocess the image
    img = Image.open(image_path)
    img = img.resize((150, 150))
    img = np.array(img) / 255.0
    img = np.expand_dims(img, axis=0)

    # Make predictions with the model
    predictions = model.predict(img)
    predicted_label = np.argmax(predictions)

    return predicted_label
```
